```python
# 직렬화/ 역직렬화

import pickle

# 직렬화할 데이터
data = {
    'name': '홍길동',
    'age': 30,
    'scores': [90, 85, 88]
}

# 직렬화하여 파일에 저장
with open('abc/data.pickle', 'wb') as f:
    pickle.dump(data, f)

# 파일에서 역직렬화하여 읽기
with open('abc/data.pickle', 'rb') as f:
    loaded_data = pickle.load(f)

print(loaded_data)  # 원본 데이터가 그대로 복원됨

```

```python
data
```

```python
import json

with open('abc/data.pickle.json', 'w') as f:
    json.dump(data, f,ensure_ascii=False)
```

```python
with open('abc/data.pickle.json', 'r') as f:
    loaded_txt = json.load(f)
```
