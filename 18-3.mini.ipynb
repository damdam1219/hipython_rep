```python
import pandas as pd

df = pd.read_csv("./data/서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv",encoding = 'cp949')
df
```

```python
# 필요없는 칼럼 삭제
df.drop(columns='작업일자',inplace = True)
```

```python
df
```

```python
df.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rcParams  # 폰트지정
rcParams['font.family'] = 'Gulim'
rcParams['axes.unicode_minus'] = False

# 출근시간 대 승차인원이 많은 역
top10 = (
    df.groupby(['호선명', '지하철역'])['07시-08시 승차인원']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# 막대차트로 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='07시-08시 승차인원', y='지하철역', data=top10, palette='viridis')
plt.title('출근 시간대 승차 인원이 많은 역 TOP 10', fontsize=14)
plt.xlabel('승차 인원 수')
plt.ylabel('역명')
plt.tight_layout()
plt.show()

# 출근시간 대 하차인원이 많은 역
top10_1 = (
    df.groupby(['호선명', '지하철역'])['08시-09시 하차인원']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# 막대차트로 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='08시-09시 하차인원', y='지하철역', data=top10_1, palette='viridis')
plt.title('출근 시간대 하차 인원이 많은 역 TOP 10', fontsize=14)
plt.xlabel('하차 인원 수')
plt.ylabel('역명')
plt.tight_layout()
plt.show()
```

```python
top10
# 이 지역에서 승차인원이 많을 것을 보아 거주지가 많을 가능성이 있다.
# 가설: 신림지역에 승차인원이 많으므로 이 지역에 거주지가 많을 가능성이 있다.
```

```python
top10_1
# 8~9 사이에 하차인원이 가장 많은 top10_1 데이터들은 회사가 많을 가능성이 있다.
# 가설: 1위인 가산디지털단지에 회사가 많을 가능성이 있다.
```

```python
# 퇴근시간 대 승차인원이 많은 역
top5 = (
    df.groupby(['호선명', '지하철역'])['18시-19시 승차인원']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# 막대차트로 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='18시-19시 승차인원', y='지하철역', data=top5, palette='viridis')
plt.title('퇴근 시간대 승차 인원이 많은 역 TOP 10', fontsize=14)
plt.xlabel('승차 인원 수')
plt.ylabel('역명')
plt.tight_layout()
plt.show()

# 퇴근시간 대 하차인원이 많은 역
top5_1 = (
    df.groupby(['호선명', '지하철역'])['19시-20시 하차인원']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# 막대차트로 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x='19시-20시 하차인원', y='지하철역', data=top5_1, palette='viridis')
plt.title('퇴근 시간대 하차 인원이 많은 역 TOP 10', fontsize=14)
plt.xlabel('하차 인원 수')
plt.ylabel('역명')
plt.tight_layout()
plt.show()
```

```python
top5
# 가산디지털단지에 퇴근시간에 승차인원이 2위로 많은 것을 보아 출근지가 많을 가능성이 높아졌다.
```

```python
top5_1
# 퇴근시간 하차인원이 신림역이 가장 많을 것을 보아 거주지가 많을 가능성이 높아졌다.
```
